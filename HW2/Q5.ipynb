```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[1], [0], [0], [1]])


def sigmoid(x):
    return 1 / (1 + np.exp(-x))


w1 = np.random.rand(2, 2)
w2 = np.random.rand(2, 1)


epochs = 10000
errors = []

for i in range(epochs):
    # فیدفوروارد
    h = sigmoid(np.dot(X, w1))
    y_pred = sigmoid(np.dot(h, w2))

    error = y - y_pred
    errors.append(np.mean(np.abs(error)))

    delta2 = error * y_pred * (1 - y_pred)
    delta1 = delta2.dot(w2.T) * h * (1 - h)
    w2 += h.T.dot(delta2)
    w1 += X.T.dot(delta1)


h = sigmoid(np.dot(X, w1))
y_pred = sigmoid(np.dot(h, w2))
print(y_pred.round())
```

```python
plt.plot(errors)
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.title('Error Plot')
plt.show()
```
